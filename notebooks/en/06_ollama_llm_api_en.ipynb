# Ollama as a local API for LLM + LangSmith context

This notebook presents an LLMOps example: serving LLM locally as an API via the ollama tool.


## Ollama

Installation and launch are system-dependent: https://ollama.com/download

`curl -fsSL https://ollama.com/install.sh | sh`

download and run LLM:

```bash
ollama pull llama3.2

```

By default, Ollama exposes the API locally at `http://localhost:11434`.

````python
import requests

payload = {
    "model": "llama3.2",
    "prompt": "Explain in 1-2 paragraphs what MLOps is.",
    "stream": False,
}

response = requests.post("http://localhost:11434/api/generate", json=payload)
print(response.json()["response"])
````

## Summary

- Classic MLOps focuses on data, experiments, models, and deployments.
- LLMOps adds prompts, traces, token costs, hallucination monitoring, security, and response quality assessment.
- Ollama is a good example of a local API serving.
- LangSmith is a good example of observability and evaluation for LLM applications.
